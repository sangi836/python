```python
import pandas as pd

# Assuming `a_df` has already been loaded as shown in your previous notebook.

# 1. Print the number of rows for the subset of the dataframe for which traffic values are higher than 10,000
high_traffic_subset = a_df[a_df['cnt'] > 10000]
num_rows = high_traffic_subset.shape[0]
print(f"Number of rows with traffic greater than 10,000: {num_rows:,}")

# 2. Print the names of all the columns of the dataframe
column_names = a_df.columns.tolist()
print("Column names in the DataFrame:")
print(column_names)

# 3. Print a subset of the dataframe containing the 10th and the 3rd rows of the original dataframe (in that order)
subset_rows = a_df.iloc[[9, 2]]  # 10th row (index 9) and 3rd row (index 2)
print("Subset containing the 10th and 3rd rows:")
print(subset_rows)

# 4. Which US state has the 8th largest total airport traffic?
state_traffic = a_df.groupby('state')['cnt'].sum().sort_values(ascending=False)
eighth_largest_state = state_traffic.index[7]  # Index 7 corresponds to the 8th largest
print(f"The state with the 8th largest total airport traffic is: {eighth_largest_state}")

# 5. Which state has the second-most total number of airports?
state_airports = a_df['state'].value_counts()
second_most_airports_state = state_airports.index[1]  # Index 1 corresponds to the second most
print(f"The state with the second-most total number of airports is: {second_most_airports_state}")

# 6. Of the cities that have multiple airports, which one has the least total traffic for all airports?
city_traffic = a_df.groupby('city')['cnt'].sum()
multi_airport_cities = a_df['city'].value_counts()[a_df['city'].value_counts() > 1].index
least_traffic_city = city_traffic[city_traffic.index.isin(multi_airport_cities)].idxmin()
least_traffic_value = city_traffic.min()
print(f"The city with multiple airports that has the least total traffic is: {least_traffic_city} with {least_traffic_value} total traffic.")

# 7. What is the total airport traffic for all states that border the state of Massachusetts?
bordering_states = ['New York', 'Connecticut', 'Rhode Island', 'New Hampshire', 'Vermont']
bordering_traffic = a_df[a_df['state'].isin(bordering_states)]['cnt'].sum()
print(f"The total airport traffic for all states that border Massachusetts is: {bordering_traffic}.")
```
